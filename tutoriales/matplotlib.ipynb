# matplotlib

Un forma simple de hacer un gráfico en Python . Importamos la librería:

```python
import matplotlib.pyplot as plt
```

el método `plot(x, y)` une con lineas los pares ordenados `(x, y)` y el método `show` nos muestra la figura

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])

plt.show()
```

a esta figura podemos editarle distintas partes, por ejemplo le agregamos otra curva y le indicamos a cada una su `label` y después mostramos la leyenda con el método `legend`

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label="una curva")
plt.plot([1, 2, 3, 4], [4, 1, 3, 2], label="otra curva")

plt.legend()
plt.show()
```

a esta misma figura podemos agregarle un título con `title`

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label="una curva")
plt.plot([1, 2, 3, 4], [4, 1, 3, 2], label="otra curva")

plt.legend()

plt.title("gráfico de ejemplo")

plt.show()
```

también podemos agregarle nombre a los ejes con `xlabel` e `ylabel`

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label="una curva")
plt.plot([1, 2, 3, 4], [4, 1, 3, 2], label="otra curva")

plt.legend()

plt.title("gráfico de ejemplo")

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

Si ambas listas se corresponden con datos experimentales y no queremos unirlos con lineas, podemos usar `scatter` para solo plotear los puntos

```python
plt.scatter([1, 2, 3, 4], [1, 4, 2, 3], label="una curva")
plt.scatter([1, 2, 3, 4], [4, 1, 3, 2], label="otra curva")

plt.legend()

plt.title("gráfico de ejemplo")

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

o si queremos usar líneas puntedas para unir los puntos le indicamos a `plot` el tipo de `marker` y de `linestyle`

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], marker="s", linestyle="dashed", label="una curva")
plt.plot([1, 2, 3, 4], [4, 1, 3, 2], marker="o", linestyle="dotted", label="otra curva")

plt.legend()

plt.title("gráfico de ejemplo")

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
